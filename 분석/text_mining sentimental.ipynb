```python
######## 데이터 불러오기 #########
import os
import nltk

# print(len(noun_adj))
# print(len(set(noun_adj)))
#pprint(nouns_tag)

def read_data(filename):
    with open(filename, 'r',encoding='utf-8') as f: #데이터 f 함수를 이용하여 불러오기
        data = [line.split('\t') for line in f.read().splitlines()] 
        #f 함수를 이용하여 먼저 줄별로 split한 데이터를 반복문을 통해서 띄어쓰기를 구분자로 split한 것을 리스트로 반환
        
        # txt 파일의 헤더(id document label)는 제외하기
        data = data[1:]
    return data
    
train_df = read_data('C:/Users/student/Documents/GitHub/Moons/분석/test3.txt')
test_df = read_data('C:/Users/student/Documents/GitHub/Moons/분석/test3.txt')
#print(train_df)


########## 토크나이즈 ###########

# 하나의 문장을 토큰화 한 후 텍스트와 품사태깅을 / 구분자로 묶어준다.
def tokenizing(file):
    return ['/'.join(t) for t in twitter.pos(file,norm=True, stem=True)]


#print(tokenizing(traub_df))
# t = tokenizing('hi. my name is jinha')
# print(t)

train_pos = [] #훈련데이터

for row in train_df:
    try:
        train_pos0 = tokenizing(row[0])
        #pprint(train_pos0)
        #리스트 안에 한문장에 대해서 위에서 만든 tokenizing함수를 통해서 [[토큰화텍스트],긍/부정 여부]를

        #리스트의 각문장별로 요소로 넣는다.

        #print(train_pos0)
        train_pos.append(train_pos0.replace)
    except:
        pass
        
test_pos = [] #테스트 데이터
for row in test_df:
    #print(row)
    try:
        test_pos0 = tokenizing(row[0])
        test_pos.append(test_pos0)
    except:
        pass
```

```python
#위에서 만든 데아터에서 긍/부정을 제외하고 token에 넣어준다. [[a],b] 에서 a만 넣는다고 생각하면 됨
tokens = [t for d in train_pos for t in d]
print(tokens)

import nltk
text = nltk.Text(tokens,name='NMSC')#nltk라이브러리를 통해서 텍스트 데이터 나열

len(set(text.tokens))#35425개의 고유 텍스트가 존재

text.vocab().most_common(10) #vocab().most_common(10) - 텍스트 빈도 상위 10개 보여주기 즉, count_values()를 통해서 내림차순한 것과 같습니다.
 
```
